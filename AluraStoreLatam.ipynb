### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# mostramos los ingresos totales de cada tienda.  .2f redondea a 2 decimales. , mejora la legibilidad agregando los separadores de miles.
print("Ingresos totales por tienda:")
print(f"Tienda 1: ${ingreso_tienda:,.2f}")
print(f"Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Tienda 4: ${ingreso_tienda4:,.2f}")

```

# 2. Ventas por categoría

```python
#  agrupamos por categoria y contamos cuántas ventas hay de cada una
categorias_t1 = tienda.groupby('Categoría del Producto')['Precio'].count().sort_values(ascending=False)
categorias_t2 = tienda2.groupby('Categoría del Producto')['Precio'].count().sort_values(ascending=False)
categorias_t3 = tienda3.groupby('Categoría del Producto')['Precio'].count().sort_values(ascending=False)
categorias_t4 = tienda4.groupby('Categoría del Producto')['Precio'].count().sort_values(ascending=False)

# mostramos resultados
print("Categorías más populares - Tienda 1:")
print(categorias_t1)
print("\nCategorías más populares - Tienda 2:")
print(categorias_t2)
print("\nCategorías más populares - Tienda 3:")
print(categorias_t3)
print("\nCategorías más populares - Tienda 4:")
print(categorias_t4)
```

# 3. Calificación promedio de la tienda


```python
#  calculamos la calificacion promedio de cada tienda
promedio_tienda1 = tienda['Calificación'].mean()
promedio_tienda2 = tienda2['Calificación'].mean()
promedio_tienda3 = tienda3['Calificación'].mean()
promedio_tienda4 = tienda4['Calificación'].mean()

#  mostramos resultados en decimales (.2f) para poder diferenciar los mejores
print("Promedio de calificaciones por tienda \n")
print(f"Tienda 1: {promedio_tienda1:.2f}")
print(f"Tienda 2: {promedio_tienda2:.2f}")
print(f"Tienda 3: {promedio_tienda3:.2f}")
print(f"Tienda 4: {promedio_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plot
ventas_t1 = tienda['Producto'].value_counts()
ventas_t2 = tienda2['Producto'].value_counts()
ventas_t3 = tienda3['Producto'].value_counts()
ventas_t4 = tienda4['Producto'].value_counts()

# mostrmos los productos mas y menos vendidos
print("Tienda 1")
print("Más vendidos:\n", ventas_t1.head(5))
print("\nMenos vendidos:\n", ventas_t1.tail(5))

print("\nTienda 2")
print("Más vendidos:\n", ventas_t2.head(5))
print("\nMenos vendidos:\n", ventas_t2.tail(5))

print("\nTienda 3")
print("Más vendidos:\n", ventas_t3.head(5))
print("\nMenos vendidos:\n", ventas_t3.tail(5))

print("\nTienda 4")
print("Más vendidos:\n", ventas_t4.head(5))
print("\nMenos vendidos:\n", ventas_t4.tail(5))

# graficamos los mas vendidos por tienda


# Tienda 1
ventas_t1.head(5).plot(kind='bar', figsize=(6,2), color='skyblue', xlabel='Ventas', ylabel= 'Productos')
plot.title('Tienda 1 - Productos más vendidos')
plot.show()

# Tienda 2
ventas_t2.head(5).plot(kind='bar', figsize=(6,2), color='green', xlabel='Ventas', ylabel= 'Productos')
plot.title('Tienda 2 - Productos más vendidos')
plot.show()

# Tienda 3
ventas_t3.head(5).plot(kind='bar', figsize=(6,2), color='purple', xlabel='Ventas', ylabel= 'Productos')
plot.title('Tienda 3 - Productos más vendidos')
plot.show()

# Tienda 4
ventas_t4.head(5).plot(kind='bar', figsize=(6,2), color='blue', xlabel='Ventas', ylabel= 'Productos')
plot.title('Tienda 4 - Productos más vendidos')
plot.show()

```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio para cada tienda
envio_prom_t1 = tienda['Costo de envío'].mean()
envio_prom_t2 = tienda2['Costo de envío'].mean()
envio_prom_t3 = tienda3['Costo de envío'].mean()
envio_prom_t4 = tienda4['Costo de envío'].mean()

#  Mostrar resultados formateados
print("Promedio de costo de envío por tienda\n")
print(f"Tienda 1: ${envio_prom_t1:,.2f}")
print(f"Tienda 2: ${envio_prom_t2:,.2f}")
print(f"Tienda 3: ${envio_prom_t3:,.2f}")
print(f"Tienda 4: ${envio_prom_t4:,.2f}")
```

# Graficos

```python
# Gráfico de barras: Ingreso total por tienda
ingresos = [
    tienda['Precio'].sum(),
    tienda2['Precio'].sum(),
    tienda3['Precio'].sum(),
    tienda4['Precio'].sum()
]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plot.figure(figsize=(8,5))
plot.bar(tiendas, ingresos, color=['skyblue','lightgreen','salmon','plum'])
plot.title("Ingreso total por tienda")
plot.xlabel("Tienda")
plot.ylabel("Ingresos")
plot.show()



# Gráfico de líneas: Costo de envío promedio por tienda
envio_promedios = [
    tienda['Costo de envío'].mean(),
    tienda2['Costo de envío'].mean(),
    tienda3['Costo de envío'].mean(),
    tienda4['Costo de envío'].mean()
]

plot.figure(figsize=(8,5))
plot.plot(tiendas, envio_promedios, marker='o', color='orange', linewidth=2)
plot.title("Costo de envío promedio por tienda")
plot.xlabel("Tienda")
plot.ylabel("Costo de envío promedio ($)")
plot.grid(True, linestyle='--', alpha=0.6)
plot.show()



promedios = [promedio_tienda1, promedio_tienda2, promedio_tienda3, promedio_tienda4]

#gráfico de área
plot.figure(figsize=(8,5))
plot.fill_between(tiendas, promedios, color='skyblue', alpha=0.5)
plot.plot(tiendas, promedios, color='blue', linewidth=2, marker='o')

plot.title("Calificación promedio por tienda", fontsize=14)
plot.xlabel("Tiendas")
plot.ylabel("Calificación promedio")
plot.ylim(0, 5)
plot.grid(True, linestyle='--', alpha=0.5)

plot.show()
```
